```python
 import pandas as pd
```

```python
# DataSets Loading
climate_raw = pd.read_csv("../data/raw/climate_data_nepal_district_wise_monthly_province_grouped.csv")
glacier_raw = pd.read_csv("../data/raw/nepal_glacier_data.csv")
```

```python
# fill the null province values with forward fill
climate_raw['province.1'] = climate_raw['province.1'].ffill()
```

```python
climate_raw.head()
```

```python
climate_raw['date'] = pd.to_datetime(climate_raw['date'], format = 'mixed', dayfirst = 'false')

```

```python
print(climate_raw['date'].dtype)
print(climate_raw['year'].dtype)
print(climate_raw['year'].dtype)
```

```python
# drop province.1
climate_raw = climate_raw.drop('province.1', axis = 1)
```

```python
# drop a duplicate value
climate_raw.drop_duplicates(inplace=True)
```

```python
glacier_raw = glacier_raw.drop(["Minimum Elevation Exposed","Mean Elevation Accumulation","Mean Elevation Ablation","Snow Line Accuracy"], axis = 1)
```

```python
# delete duplicate values
glacier_raw.drop_duplicates(inplace=True)
```

```python
# save the processed data 
climate_raw.to_csv('../data/processed/climate_data_cleaned.csv', index=False)
glacier_raw.to_csv('../data/processed/glacier_data_cleaned.csv', index = False)
```
